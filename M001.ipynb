## Navigation mit Zellen
- Enter -> in Zelle gehen
- Escape -> Zelle verlassen
- Wenn Zelle verlassen
    - Pfeiltasten navigieren
    - M -> MarkDown(Text) Zelle
    - Y -> Code Zelle
- Wenn in Zelle
    - Schreiben
    - Shift+Enter -> Zelle Ausführen & Neue Zelle darunter
