## Globaler vs. Objektorientierter Modus
Globaler Modus: Befehle ausführen über plt, mit plt.show() Konfiguration abschließen

Objektorientierter Modus: Figure Objekte und Axis Objekte werden didiziert angesprochen und dadurch konfiguriert

```python
plt.plot(pop["Population"])
plt.grid(alpha = 0.2)
plt.show() #  Schließt die Konfiguration und zeigt ihn an

plt.plot(pop["WorldSharePct"])
plt.grid(alpha = 0.2)
plt.show()
```

```python
plt.figure(figsize=(12,3))
plt.plot(pop["Population"])
plt.grid(alpha = 0.2)
plt.show()
```

```python
fig, ((ax1_1, ax1_2), (ax2_1, ax2_2)) = plt.subplots(2,2)
ax1_1.plot(np.random.random(100))
ax2_1.pie(pop["WorldSharePct"])
ax1_2.bar(pop.index, pop["Population"])
ax2_2.hist(np.random.randint(100, size = 100), bins = 20)
fig.savefig("data/Test.png")
```

```python
plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
