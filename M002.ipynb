### Performance

```python
import time
def measureTime(func):
    start = time.time()
    func()
    end = time.time()
    print(end-start)
```

```python
anz = 100_000_000
```

```python
def pythonList():
    x = list(range(anz))
    del(x)
```

```python
def numpyArray():
    x = np.array(range(anz))
    del(x)
```

```python
def numpyArange():
    x = np.arange(anz)
    del(x)
```

```python
measureTime(pythonList)
measureTime(numpyArray)
measureTime(numpyArange)
```

##### hstack, vstack
