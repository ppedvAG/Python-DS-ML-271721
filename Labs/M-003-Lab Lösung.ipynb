Gib die Anzahl der Datensätze pro Altersgruppe an

```python
data.groupby(data["Age_Group"]).count()
```

Welche unserer Produkte bringen im Durchschnitt den höchsten Profit ein?

```python
data.groupby(data["Product"])["Profit"].mean().sort_values(ascending=False)
```

Finde die Stadt + Land Kombination mit dem höchsten durchschnittlichen Profit.

```python
data.groupby(data["Country"] + ", " + data["State"])["Profit"].mean().sort_values(ascending=False)
```
