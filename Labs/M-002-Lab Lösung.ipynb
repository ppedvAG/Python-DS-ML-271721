```python
import numpy as np
```

## 1. Erstellung von Arrays
Erstelle ein Array mit Zahlen von -10 bis (einschließlich) 10

```python
np.arange(-10, 11)
```

Erstelle ein 5x5 Array mit den Zahlen von 0 bis 24

```python
np.arange(25).reshape(5, 5)
```

Erstelle ein Array mit ungeraden Zahlen von 0 bis 100

```python
np.arange(1, 101, 2)
```

Erstelle ein Array befüllt mit Kommazahlen von 0 bis 10 (Schrittgröße 0.1)

```python
np.arange(0, 10, 0.1)
```

Erstelle ein Array von 50 bis 100 und summiere alle Zahlen

```python
np.arange(50, 101).sum()
```

Erstelle ein 5x10 Array und berechne den Mittelwert jeder Zeile

```python
np.random.randint(10, size=(5, 10)).mean(axis=1)
```

Erstelle ein 5x10 Array und berechne die Summe jeder Spalte

```python
np.random.randint(10, size=(5, 10)).sum(axis=0)
```

## 2. Vektorisierung

Erzeuge ein Array mit zufälligen Kommazahlen von 0 bis 1 und finde alle Elemente die mindestens 0.5 sind

```python
x = np.random.random(10)
x[x > 0.5]
```

Erzeuge ein Array der Form x^2 von 1 bis 20
 - Finde nun in diesem Array alle Zahlen die gerade sind

```python
np.arange(1, 21) ** 2
```

```python
y = np.arange(1, 21) ** 2
y[y % 2 == 0]
```

Erzeuge ein Array mit 10 Stellen befüllt mit ganzen Zufallszahlen von 0 bis 100. Finde alle Elemente die maximal 80% des Höchstwertes sind.

```python
z = np.random.randint(100, size=10)
m = z.max()
m
```

```python
z[z > (m * 0.8)]
```

Erzeuge ein 10x10 Array befüllt mit ganzen Zufallszahlen von 0 bis 10. Gib nun alle Zeilen aus, bei denen der neue Mittelwert der entsprechenden Zeile größer als 5 ist.

```python
a = np.random.randint(10, size=(10, 10))
```

```python
a.mean(axis=1)
```

```python
a[a.mean(axis=1) > 5]
```

Erzeuge zwei Arrays mit gleicher Länge und Zufallszahlen. Finde alle Werte, die in beiden Arrays an derselben Stelle sind
- Finde zusätzlich die Stellen der gleichen Zahlen

```python
b = np.random.randint(10, size=100)
c = np.random.randint(10, size=100)
```

```python
b == c
```

```python
# Gleiche Werte
b[b == c]
```

```python
# Stellen
np.arange(100)[b == c]
```

```python
# Beides gleichzeitig
maske = b == c
np.vstack((np.arange(100)[maske], b[maske], c[maske]))
```
