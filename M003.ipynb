## Serie
Ein Numpy Array mit extra Eigenschaften

Eigenschaften
- Name
- Benannter Index
- ...

Auf das unterliegende Numpy Array kann immer zugegriffen werden
-> alle Numpy Array Funktionen können auf eine Serie angewendet werden

```python
pd.Series([8.6,5.6,1.2,0.2])
```

```python
e = pd.Series([8.6, 83.7, 0.38, 9])
```

### read_csv
Daten aus einer CSV-Quelle einlesen

Wird per Parameter konfiguriert

Beispiele:
- delimiter
- thousands
- decimal
- parse_dates
- index_col
- ...

https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

```python
data =  pd.read_csv("data/PopulationData.csv", delimiter=";", thousands=",", decimal=".")
```

```python
data
```

```python
data.info()
```

#### Probleme mit dem Datenset

DataFrame:
- Schlechte Spaltennamen
- Doppelter Index

Daten selbst:
- Prozentzeichen (str nicht float)
- N.A.
- NaN
- Datentypen

##### Index setzen
set_index("Name") ODER index_col bei read_csv(...)

WICHTIG: Jeder Befehl erzeugt immer nur eine Kopie des Datensets (mit der Änderung)

Zwei Optionen:
- Bestehende Variable überschreiben
- inplace = true

```python
#data = data.set_index("Country (or dependency)") # Möglichkeit 1
data.set_index("Country (or dependency)", inplace=True) # Möglichkeit 2
```

```python
data
```

```python
data.index.name = "Country"
```

```python
data
```

##### Spalten anpassen
Spalten entfernen: drop(...)

Spalten umbenennen: rename(...)

rename benötigt ein Dictionary mit den columns = Parameter

```python
data.drop(columns=["#"], inplace=True)
```

```python
data
```

```python
data.rename(columns= {
    "Population(2020)": "Population",
    "Density(P/Km²)": "Density",
    "Land Area(Km²)": "LandArea",
    "Migrants(net)": "Migrants",
    "Fert.Rate": "FertRate",
    "Med.Age": "MedAge",
    "UrbanPop %": "UrbanPopPct",
    "WorldShare": "WorldSharePct",
}, inplace=True)
```

```python
data
```

#### Daten Speichern

```python
data.to_csv("data/PopulationDataFastFertig.csv")
```

#### Sortieren
sort_index()

alle Sortierungen können auch mit ascending = True/False angegeben werden (auf-/absteigend)

```python
data.sort_index()
```

#### Auf Daten zugreifen
Drei Möglichkeiten
- Index: Zugriff auf Zeilen/Spalten/Kombination
- loc: Mehrere Spalten
- iloc: Wie loc, aber nur mit Zahlen

```python
data["Population"] # Spalte über Index
```

```python
data["China": "Russia"] # Index von Zeilen
```

```python
data["Population"]["China"]
```

```python
data["Population"]["China":"Russia"]
```

```python
data.loc["China":"Russia", "Population":"Density"]
```

```python
data.loc["China":"Russia", ("Population","Density")] # Einzelne Spalten Selektieren mit ("..","..")
```

```python
data.loc[("China","Russia"), :] # Einzelne Zeilen Selektieren mit ("..","..")
```

```python
data.iloc[10:20,0:3]
```

### Daten filtern

```python
data[data["Population"]>50_000_000]
```
