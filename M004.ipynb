## Prozentzeichen
Muss mithilfe von String Replace gemacht werden

```python
data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data["YearlyChange"]=data["YearlyChange"].replace(" %", "", regex=True)
```

```python
data
```

```python
def fixPct(colName):
    x = data[colName].replace(" %", "", regex=True)
    data[colName]=x
```

```python
for col in data.columns:
    fixPct(col)
```

```python
data
```
